#Reading Data

```python
import pandas as pd
```

#Load Data

```python
data = pd.read_csv('/content/drive/MyDrive/Data Science/Sleep_health_and_lifestyle_dataset.csv')
data.sample(10)
```

#Data Preprocessing

```python
def numb(c):
  if c == 'Normal':
    return 2
  elif c == 'Normal Weight':
    return 2
  elif c == 'Overweight':
    return 1
  elif c == 'Obese':
    return 0
  else:
    return 0
```

```python
data['BMI Category'] = data['BMI Category'].apply(numb)
```

```python
data[['High Pressure', 'Low Pressure']] = data['Blood Pressure'].str.split('/', expand=True)
```

```python
data = data.dropna()
```

```python
def converttype(n):
  n = int(n)
  return n
```

```python
data['High Pressure'] = data['High Pressure'].apply(converttype)
```

```python
data['Low Pressure'] = data['Low Pressure'].apply(converttype)
```

```python
data = data.drop('Blood Pressure', axis=1)
```

```python
data.head()
```

```python
data.head()
```

```python
data = data.drop(['Person ID' , 'Sleep Duration','Stress Level','Physical Activity Level'], axis = 1)
```

```python
data.head()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
def transform(m):
    if(m=='None'):
      return 0
    elif(m=='Sleep Apnea'):
      return 1
    elif(m=='Insomnia'):
      return 2
```

```python
data['Sleep Disorder'] = data['Sleep Disorder'].apply(transform)
```

```python
data = data.drop(['Gender'], axis = 1)
```

```python
data.head()
```

```python
y = data['Sleep Disorder']
```

```python
y.value_counts()
```

```python
y
```

```python
x = data.drop(['Sleep Disorder'], axis = 1)
```

```python
x = pd.get_dummies(x, drop_first=True)
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

# ML MODELS (Algorithms)


```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

```python
hyperparameters = {'kernel':('linear', 'rbf'), 'C':[1, 10]}
model = SVC()
clf = GridSearchCV(model, hyperparameters)
clf.fit(X_train, y_train)
```

```python
print(clf.best_params_)
```

```python
model = SVC(C = 10, kernel =  'rbf')
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import f1_score
y_pred = model.predict(X_test)
f1 = f1_score(y_test, y_pred, average=None)
print(f1)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=19)
model.fit(X_train, y_train)
model.score(X_test, y_test)
y_pred = model.predict(X_test)
f1 = f1_score(y_test, y_pred, average=None)
print(f1)
```

```python
from xgboost import XGBClassifier
```

```python
model3 = XGBClassifier()
model3.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = model3.predict(X_test)
print(classification_report(y_test, y_pred))
```
